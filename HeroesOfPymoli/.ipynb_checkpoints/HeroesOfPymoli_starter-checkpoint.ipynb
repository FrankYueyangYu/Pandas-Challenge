```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
smry_list=[[purchase_data['Item ID'].nunique(),purchase_data['Price'].mean(),
           purchase_data['Purchase ID'].count(),purchase_data['Price'].sum()]]
smry_total=pd.DataFrame(smry_list)
smry_total.columns=['Number_of_Unique_Items','Average_Price','Number_of_Purchases','Total_Revenue']
smry_total['Average_Price']=smry_total['Average_Price'].round(2)
smry_total
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
unique_player=purchase_data.drop_duplicates('SN')
unique_player.shape
```

```python

gender_demo=pd.DataFrame([purchase_data.groupby('SN').first()['Gender'].value_counts(),
             purchase_data.groupby('SN').first()['Gender'].value_counts(normalize=True)]).transpose()
gender_demo.columns=['Total_Count','Percentage_of_Players']
gender_demo['Total_Count']=gender_demo['Total_Count'].astype(int)
gender_demo['Percentage_of_Players']=(gender_demo['Percentage_of_Players']*100).round(2).map("{:.2f}%".format)
gender_demo
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data.groupby('Gender')['Price'].sum()/purchase_data['Gender'].value_counts().sort_index()
```

```python
gender_purchase=pd.DataFrame([purchase_data['Gender'].value_counts().sort_index(),
                 purchase_data.groupby('Gender')['Price'].mean(),
                 purchase_data.groupby('Gender')['Price'].sum(),
                 purchase_data.groupby('Gender')['Price'].sum()/purchase_data['Gender'].value_counts().sort_index()]).transpose()
                             
gender_purchase.columns=['Purchase_Count','Average_Price','Total_Purchase','Average_Purchase_per_Person']
gender_purchase['Purchase_Count']=gender_purchase['Purchase_Count'].astype(int)
gender_purchase.style.format({'Average_Price':'${:.2f}'.format,
                              'Total_Purchase':'${:.2f}'.format,
                              'Average_Purchase_per_Person':'${:.2f}'.format          
                             })
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
purchase_data['Age'].describe()
```

```python
bins=[0,10,15,20,25,30,35,40,purchase_data['Age'].max()+1]
labels=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
age_demo=pd.DataFrame([unique_player.groupby(pd.cut(unique_player['Age'],bins=bins,labels=labels,right=False)).size(),
                       unique_player.groupby(pd.cut(unique_player['Age'],bins=bins,labels=labels,right=False)).size()/780]).transpose()
age_demo.columns=['Total_Count','Percentage_of_Players']
age_demo['Total_Count']=age_demo['Total_Count'].astype(int)
age_demo['Percentage_of_Players']=(age_demo['Percentage_of_Players']*100).map("{:.2f}%".format)
age_demo
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_purchase=pd.DataFrame([purchase_data.groupby(pd.cut(purchase_data['Age'],bins=bins,labels=labels,right=False)).size(),
                           purchase_data.groupby(pd.cut(purchase_data['Age'],bins=bins,labels=labels,right=False))['Price'].mean(),
                           purchase_data.groupby(pd.cut(purchase_data['Age'],bins=bins,labels=labels,right=False))['Price'].sum(),
                           purchase_data.groupby(pd.cut(purchase_data['Age'],bins=bins,labels=labels,right=False))['Price'].sum()/
                           unique_player.groupby(pd.cut(unique_player['Age'],bins=bins,labels=labels,right=False)).size()]).transpose()
age_purchase.columns=['Purchase_Count','Avg_Price','Total_Purchase','Avg_per_Person']
age_purchase['Purchase_Count']=age_purchase['Purchase_Count'].astype(int)

age_purchase['Avg_Price']=age_purchase['Avg_Price'].map("${:.2f}".format)
age_purchase['Total_Purchase']=age_purchase['Total_Purchase'].map("${:.2f}".format)
age_purchase['Avg_per_Person']=age_purchase['Avg_per_Person'].map("${:.2f}".format)
age_purchase
                                                 
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
whales=pd.DataFrame([purchase_data.groupby("SN").size(),
                     purchase_data.groupby("SN")['Price'].mean(),
                     purchase_data.groupby("SN")['Price'].sum()]).transpose()

whales.columns=['Total_Count','Avg_Price','Total_Value']
whales.sort_values(by='Total_Value',ascending=False,inplace=True)
whales['Total_Count']=whales['Total_Count'].astype(int)
whales.style.format({'Avg_Price':'${:.2f}'.format,
                    'Total_Value':'${:.2f}'.format})
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
unique_items=purchase_data.drop_duplicates('Item ID')
unique_items=unique_items.filter(items=['Item ID','Item Name'])
items_group=pd.DataFrame([purchase_data.groupby('Item ID').size(),
                          purchase_data.groupby('Item ID')['Price'].mean(),
                          purchase_data.groupby('Item ID')['Price'].sum()]).transpose()
items_group.columns=['Count','Avg_Price','Total_Value']
popular_items=pd.merge(unique_items,items_group,on='Item ID')
popular_items.sort_values(by='Count',ascending=False,inplace=True)
popular_items['Count']=popular_items['Count'].astype(int)
popular_items.style.format({'Avg_Price':'${:.2f}'.format,'Total_Value':'${:.2f}'.format})
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
profit_items=popular_items.sort_values(by='Total_Value',ascending=False)
profit_items.head()
```
